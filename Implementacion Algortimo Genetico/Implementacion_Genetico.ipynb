```python
#Subiendo archivos, almacenadolos y calculando las coordenadas geograficas en radianes
dataframe1 = pd.read_csv('RUTAS_-_Hoja_8.csv',index_col=0)
nomdepartamento = dataframe1.index[0]
coord_geo = np.array(list(zip(dataframe1.LATITUD,dataframe1.LONGITUD)))
coord_geo_rad=coord_geo.copy()

coord_geo_lat = np.array(list(zip(dataframe1.LATITUD)))
coord_geo_lon = np.array(list(zip(dataframe1.LONGITUD)))


for i in range(len(coord_geo_rad)):
  for j in range(len(coord_geo_rad[i])):
    coord_geo_rad[i][j]= math.radians(coord_geo_rad[i][j])

dataframe1['LATITUD_RAD']=coord_geo_rad[:,0]
dataframe1['LONGITUD_RAD']=coord_geo_rad[:,1]
dataframe1
```
